## Memperbaiki Kualitas Data

- Mari kita berikan faktor tambahan pada data untuk mempermudah analisis.

```python
# Tambahkan nilai waktu dan tanggal pada saat iklan ditayangkan
vehicles_data['posting_date'] = vehicles_data['date_posted'].dt.date
vehicles_data['posting_time'] = vehicles_data['date_posted'].dt.time
```

- Setelah diteliti kembali, kita akan membutuh kan kolom date_posted. Hal ini nanti akan berkaitan dengan pengiklanan pada proses bisnis yang ada di perusahaan yang berarti akan berpengaruh terhadap beberapa kolom. Setelah langkah-langkah di atas, kita coba buat dua kolom baru, yaitu 'posting_date' yang berisi tanggal iklan dan 'posting_time' yang berisi waktu iklan. Hal ini dapat menggunakan kolom-kolom ini untuk analisis lebih lanjut berdasarkan waktu dan tanggal iklan ditayangkan.

```python
# Tambahkan usia kendaraan saat iklan ditayangkan
vehicles_data['vehicle_age'] = vehicles_data['date_posted'].dt.year - vehicles_data['model_year']
```

- Kita akan gunakan kolom 'model_year' untuk menghitung usia kendaraan saat iklan ditayangkan. Usia kendaraan dapat dihitung dengan mengurangkan tahun model dari tahun iklan ditayangkan.

```python
# Tambahkan jarak tempuh rata-rata kendaraan per tahun 
average_odometer_by_year = vehicles_data.groupby('model_year')['odometer'].mean()
vehicles_data['average_odometer_by_year'] = vehicles_data['odometer'] / vehicles_data['vehicle_age']
vehicles_data['average_odometer_by_year'].fillna(0, inplace=True)
```

- Kita akan menggunakan metode .groupby() untuk mengelompokkan data berdasarkan tahun model kendaraan, dan kemudian hitung rata-rata jarak tempuh untuk setiap grup.

```python
# Mungkin membantu untuk mengganti nilai pada kolom 'condition' dengan sesuatu yang bisa dimanipulasi lebih mudah
condition_mapping = {
    'new': 5,
    'like new': 4,
    'excellent': 3,
    'good': 2,
    'fair': 1,
    'salvage': 0
}
vehicles_data['condition'] = vehicles_data['condition'].map(condition_mapping)
```

- Dengan dictionary yang kita buat. Kita akan mudah untuk mapping kondisi setiap mobil yang berada di dataset hanya dengan menggunakan angka. Hal ini untuk mempermudah manipulasi data.

## Mempelajari dan Menangani Outlier
Berdasarkan hasil sebelumnya, kita akan tentukan kolom apa yang mungkin berisi outliers dan membuat sebuah DataFrame tanpa outliers tersebut. Outliers bisa diidentifikasi dengan menentukan batas bawah/atas untuk rentang nilai normal,

- Mari kita tentukan outliers pada kolom yang kita tentukan kita akan memilih



```python
def calculate_outlier_bounds(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_whisker = Q1 - (1.5 * IQR)
    upper_whisker = Q3 + (1.5 * IQR)
    return lower_whisker, upper_whisker


```

- Untuk data yang tidak akan kita gunakan yaitu cylinders dan condition, karena boxplot kurang cocok untuk tipe data kategori bersifat bertingkat.

----------------------------

```python
# Tentukan batas atas outlier & Tentukan batas bawah outlier pada kolom price
plt.figure(figsize=(10, 5))
sns.boxplot(x=vehicles_data['price'], color='green')
plt.title('Box Plot Harga Kendaraan')
plt.xlabel('Harga')
plt.show()

lower_price, upper_price = calculate_outlier_bounds(vehicles_data, 'price')
print("Lower Whisker, Upper Whisker")
print()
print(lower_price,"     ", upper_price)
```

```python
vehicle_type_counts = vehicles_data['type'].value_counts()

# Memilih jenis kendaraan populer: SUV, truck, sedan
popular_vehicle_types = ['SUV', 'truck', 'sedan']

# Looping untuk masing-masing jenis kendaraan populer
for vehicle_type in popular_vehicle_types:
    print(f"Analisis untuk Jenis Kendaraan: {vehicle_type}")
    
    # Mengambil data untuk jenis kendaraan tertentu
    vehicle_data = vehicles_data[vehicles_data['type'] == vehicle_type]
    
    # Analisis usia kendaraan dengan histogram
    plt.figure(figsize=(10, 5))
    sns.histplot(data=vehicle_data, x='vehicle_age', bins=20, kde=False, color='blue')
    plt.title(f'Histogram Usia Kendaraan ({vehicle_type})')
    plt.xlabel('Usia Kendaraan')
    plt.ylabel('Frekuensi')
    plt.show()
    
    # Analisis jarak tempuh dengan histogram
    plt.figure(figsize=(10, 5))
    sns.histplot(data=vehicle_data, x='odometer', bins=20, kde=False, color='green')
    plt.title(f'Histogram Jarak Tempuh ({vehicle_type})')
    plt.xlabel('Jarak Tempuh')
    plt.ylabel('Frekuensi')
    plt.show()
    
    # Analisis kondisi kendaraan dengan histogram
    plt.figure(figsize=(10, 5))
    sns.histplot(data=vehicle_data, x='condition', bins=5, kde=False, color='purple')
    plt.title(f'Histogram Kondisi Kendaraan ({vehicle_type})')
    plt.xlabel('Kondisi Kendaraan')
    plt.ylabel('Frekuensi')
    plt.show()
    
    print("\n")
```
